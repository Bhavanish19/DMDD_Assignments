```python
url = 'https://www.swiggy.com/city/mumbai/top-rated-collection?page=4'
page = requests.get(url).text
print(page)
```

```python
rating = soup.find_all('div',class_="_9uwBC wY0my")
for i in rating:
    List_rating.append(i.text)
len(List_rating)
```

```python
rating = soup.find_all('div',class_="_9uwBC _2lAlc")
for i in rating:
    List_rating.append(i.text)
len(List_rating)
```

```python
Avgcost = soup.find_all('div',class_="nVWSi")
for i in Avgcost:
    List_Avgcost.append(i.text)
len(List_Avgcost)
```

```python
Speciality = soup.find_all('div',class_="_1gURR")
for i in Speciality:
    List_Speciality.append(i.text)
len(List_Speciality)
```

```python
timetaken = soup.find_all('div',class_="_3Mn31")
for i in timetaken:
    List_Timetaken.append(i.text)
len(List_Timetaken)
```

```python
# data cleaning only taking the 5 and 6 index of List_Timetaken to extract time taken
List_Timetaken1=[]
for i in range(len(List_Timetaken)):
    List_Timetaken1.append(List_Timetaken[i][4:6])
#List_Timetaken1.append
List_Timetaken1
```

```python
from uuid import uuid4
List_Order_id=[]
for i in List_Rest_Name:
    List_Order_id.append(str(uuid4()))
    
```

```python
List_City=[]
```

```python
for i in range (len(List_Timetaken1)):
    List_City.append('Mumbai')
```

```python
data ={'Order_ID':List_Order_id,'Restaurent':List_Rest_Name,'Speciality':List_Speciality,'Rating':List_rating,'Av_cost':List_Avgcost,'Timetaken':List_Timetaken1,'City':List_City}
```

```python
df = pd.DataFrame(data)
df
```

```python
#used to clean special character like ₹ in dataframe column Avgcost
df['Av_cost'] = df['Av_cost'].str.replace('\W', '', regex=True)
```

```python
df
#df["TimeTaken"].replace("57Mins", "Mins")
```

```python
#used to remove FORTWO letters in Avgcost Column in DataFrame
df = df.replace("FORTWO"," ",regex = True)
```

```python
df
```

```python
#delete empty rows 
new_df = df.dropna()
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
new_df
```
