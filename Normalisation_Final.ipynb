```python
# Creating a function for repeating value in list
def RepeatList(New_List,Old_List):
    for i in range(113) :
        New_List.append(Old_List[i])
        New_List.append(Old_List[i])
```

```python
List_OrderIDN=[]
List_DeliverypersonAgeN=[]
List_DeliverypersonRatingsN=[]
List_RestaurantNameN=[]
List_Restaurent_idN=[]
List_CityN=[]
List_AddressN=[]
List_OrderDateN=[]
List_TimeOrderdN=[]
List_TimeOrderpickedN=[]
List_ClimateN=[]
List_RoadtrafficdensityN=[]
List_TypeofvehicleN=[]
List_TimetakenN=[]
List_DeliveryIDN=[]
List_AvcostN=[]
List_AgentFirstNameN=[]
List_AgentLastNameN=[]
```

```python
RepeatList(List_OrderIDN,List_OrderID)
RepeatList(List_DeliverypersonAgeN,List_DeliverypersonAge)   
RepeatList(List_DeliverypersonRatingsN,List_DeliverypersonRatings) 
RepeatList(List_RestaurantNameN,List_RestaurantName) 
RepeatList(List_Restaurent_idN,List_Restaurent_id)   
RepeatList(List_CityN,List_City)  
RepeatList(List_AddressN,List_Address)  
RepeatList(List_OrderDateN,List_OrderDate)   
RepeatList(List_TimeOrderdN,List_TimeOrderd)   
RepeatList(List_TimeOrderdN,List_TimeOrderd)     
RepeatList(List_TimeOrderpickedN,List_TimeOrderpicked)   
RepeatList(List_ClimateN,List_Climate)  
RepeatList(List_RoadtrafficdensityN,List_Roadtrafficdensity)   
RepeatList(List_TypeofvehicleN,List_Typeofvehicle) 
RepeatList(List_TimetakenN,List_Timetaken)  
RepeatList(List_DeliveryIDN,List_DeliveryID) 
RepeatList(List_AvcostN,List_Avcost)
RepeatList(List_AgentFirstNameN,List_AgentFirstName) 
RepeatList(List_AgentLastNameN,List_AgentLastName)
```
